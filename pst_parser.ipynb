```python
import pypff
pst = pypff.file()
pst.open("C:\\list\\archive.pst")
```

```python
root = pst.get_root_folder()
```

```python
def parse_folder(base):
    messages = []
    a = 1
    for folder in base.sub_folders:
        if folder.number_of_sub_folders:
            messages += parse_folder(folder)
        print(folder.name)
        #msg = folder.get_sub_item(0)
        
        #with open("c:\\list\\body\\"+ str(a) +".msg",'wb') as f:
        #    f.write(msg)
        #    a = a + 1
        
        for message in folder.sub_messages:
            print(message.number_of_attachments)

            messages.append({
                "subject": message.subject,
                "sender": message.sender_name,
                "datetime": message.client_submit_time,
                "body": message.plain_text_body,
                "attachments": message.attachments,
                "number_of_attachments": message.number_of_attachments
            })
            print(message.subject)
            print(message.plain_text_body.decode('utf-8'))
            #print(message)

    return messages

messages = parse_folder(root)
```

```python
import pandas as pd
df = pd.DataFrame(messages)
```
